# Visualizing The Gender Gap In College Degrees

## Introduction

[The Department of Education Statistics](http://nces.ed.gov/programs/digest/2013menu_tables.asp) releases a data set annually containing the percentage of bachelor's degrees granted to women from 1970 to 2012. The data set is broken up into 17 categories of degrees, with each column as a separate category.

In this Project we will gain some insights about our Data using visualizations.

Tell more fast through visualization about unequal percentage between men and women across different degrees.

We have seventeen degrees in our set, we will start with STEM degrees.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

we will categorize our data into three categories:
>* STEM
* liberal arts
* other

where our first column reflects STEM degrees respectively.

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
fig = plt.figure(figsize=(18, 20))
cb_light_gray= c=(171/255, 171/255, 171/255)
#first column STEM degrees
for sp in range(0,17,3):
    ax = fig.add_subplot(6,3,sp+1)
    deg_index= int((sp)/3)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[deg_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[deg_index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=cb_light_gray, alpha=0.)
    ax.set_title(stem_cats[deg_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    if deg_index == 0:
        ax.text(2005, 15, 'Men')
        ax.text(2002, 82, 'Women')
    elif deg_index == 5:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
        ax.tick_params(labelbottom='on')
        
#first column liberal arts degrees
for sp in range(1,14,3):
    ax = fig.add_subplot(6,3,sp+1)
    deg_index= int((sp-1)/3)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[deg_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[deg_index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=cb_light_gray, alpha=0.4)
    ax.set_title(lib_arts_cats[deg_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    if deg_index == 0:
        ax.text(2005, 22, 'Men')
        ax.text(2002, 82, 'Women')
    elif deg_index == 4:
        ax.tick_params(labelbottom='on')        
#first column other degrees
for sp in range(2,18,3):
    ax = fig.add_subplot(6,3,sp+1)
    deg_index= int((sp-1)/3)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[deg_index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[deg_index]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c=cb_light_gray, alpha=0.4)
    ax.set_title(other_cats[deg_index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    if deg_index == 0:
        ax.text(2005, 7, 'Men')
        ax.text(2002, 90, 'Women')
    elif deg_index == 5:
        ax.text(2005, 64, 'Men')
        ax.text(2001, 32, 'Women')
        ax.tick_params(labelbottom='on')
#Export the plot to a png file
plt.savefig('gender_degrees.png')
plt.show()
```
